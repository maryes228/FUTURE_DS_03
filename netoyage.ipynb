```python
import os
import pandas as pd
import zipfile

# 1. Décompresser le fichier ZIP dans ./data
zip_path = "./ga-customer-revenue-prediction.zip"
extract_path = "./data"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Fichiers extraits dans :", extract_path)

# 2. Charger les fichiers train.csv et test.csv
train = pd.read_csv(os.path.join(extract_path, "train.csv"))
test = pd.read_csv(os.path.join(extract_path, "test.csv"))

print("Train shape:", train.shape)
print("Test shape:", test.shape)

# 3. Supprimer les colonnes techniques inutiles
cols_to_drop = ["visitId", "sessionId", "visitNumber"]
train_clean = train.drop(columns=cols_to_drop, errors="ignore")
test_clean = test.drop(columns=cols_to_drop, errors="ignore")

# 4. Convertir les dates au bon format
train_clean["date"] = pd.to_datetime(train_clean["date"], format="%Y%m%d")
test_clean["date"] = pd.to_datetime(test_clean["date"], format="%Y%m%d")

# 5. Remplacer les valeurs manquantes par 0
train_clean = train_clean.fillna(0)
test_clean = test_clean.fillna(0)

# 6. Sauvegarder les fichiers nettoyés en local
train_clean.to_csv("./data/train_clean.csv", index=False)
test_clean.to_csv("./data/test_clean.csv", index=False)

print("Prétraitement terminé. Fichiers enregistrés : train_clean.csv et test_clean.csv")
```
